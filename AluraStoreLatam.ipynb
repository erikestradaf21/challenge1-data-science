### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
import matplotlib.pyplot as plt

nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
tiendas_df = [tienda, tienda2, tienda3, tienda4]

# Calcula e imprime los ingresos de cada tienda.
ingresos = []
for nombre, df in zip(nombres, tiendas_df):
    ingreso = df["Precio"].sum()
    ingresos.append(ingreso)
    print(f"Ingreso total de {nombre}: $ {ingreso:,.2f}")
```

```python
# Genera un gráfico de barras con los ingresos por tienda (Visualización).
def graficar_ingresos(nombres: list[str], ingresos: list[float]) -> None:
    plt.figure(figsize=(8, 5))
    plt.bar(nombres, ingresos)
    plt.title("Ingreso Total por Tienda")
    plt.ylabel("Ingreso ($)")
    plt.xlabel("Tienda")

graficar_ingresos(nombres, ingresos)
```

# 2. Ventas por categoría

```python
"""
En este debes calcular la cantidad de productos vendidos por categoría en cada tienda.
La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo,
mostrando las categorías más populares de cada tienda.
"""
def mostrar_ventas_por_categoria(nombre, tiendas_df):
    # Conteo de ventas por categoría
    conteo = tiendas_df['Categoría del Producto'].value_counts()
    conteo.plot.pie(
        autopct='%1.1f%%',
        startangle=90,
        legend=False,
        ylabel=''
    )
    plt.title(f"Ventas por Categoría en {nombre}")
    plt.tight_layout()
    plt.show()

# Calcula e imprime las ventas por categoría
for nombre, df in zip(nombres, tiendas_df):
    print(f"\nLas 3 Categorías más vendidas en {nombre}:\n")
    print(df['Categoría del Producto'].value_counts().head(3))
    mostrar_ventas_por_categoria(nombre, df)
```
